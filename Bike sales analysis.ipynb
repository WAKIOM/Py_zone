```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the data from csv and print the first row and see how the data looks like
data = pd.read_csv(r"C:\Users\HP\Desktop\STUFF\DATA\Kaggle datasets\Sales.csv")
first_row = data.head(1)
print(first_row)
```

```python
#What product category generates the most profit
plt.bar(data['Product_Category'], data['Profit'], color='Orange')
plt.xlabel('Product_Category')
plt.ylabel('Profit')
plt.yticks([2000, 4000, 6000, 8000, 10000, 12000, 14000], ['2k', '4k', '6k', '8k','10k','12k','14k'])
plt.title('Profits by Product categories')
plt.show()
```

```python
# Group the data by 'Age_Group' and calculate the sum of 'Profit'
grouped_data = data.groupby('Age_Group')['Profit'].sum()

# Get the values and labels for the pie chart
values = grouped_data.values
labels = grouped_data.index
plt.title('Profits by age group')

# Plot the pie chart
plt.pie(values, labels=labels)
plt.show()
```
